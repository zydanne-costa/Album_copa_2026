```python
# =============================================================================
# 01. IMPORTAÇÃO DE BIBLIOTECAS
# Responsável por importar as bibliotecas utilizadas no projeto.
# =============================================================================

import pandas as pd
```

```python
# =============================================================================
# 02. ESTRUTURA DOS GRUPOS DA COPA DO MUNDO 2026
# Armazena as seleções participantes organizadas por grupo.
# =============================================================================

grupos = {
    "A": ["México", "África do Sul", "Coreia do Sul", "Tchéquia"],
    "B": ["Canadá", "Bósnia e Herzegovina", "Catar", "Suíça"],
    "C": ["Brasil", "Marrocos", "Haiti", "Escócia"],
    "D": ["Estados Unidos", "Paraguai", "Austrália", "Turquia"],
    "E": ["Alemanha", "Curaçao", "Costa do Marfim", "Equador"],
    "F": ["Holanda", "Japão", "Suécia", "Tunísia"],
    "G": ["Bélgica", "Egito", "Irã", "Nova Zelândia"],
    "H": ["Espanha", "Cabo Verde", "Arábia Saudita", "Uruguai"],
    "I": ["França", "Senegal", "Iraque", "Noruega"],
    "J": ["Argentina", "Argélia", "Áustria", "Jordânia"],
    "K": ["Portugal", "RD do Congo", "Uzbequistão", "Colômbia"],
    "L": ["Inglaterra", "Croácia", "Gana", "Panamá"]
}
```

```python
# =============================================================================
# 03. SIGLAS FIFA
# Relaciona cada seleção ao seu código FIFA de três letras.
# =============================================================================

siglas = {
    "México": "MEX",
    "África do Sul": "RSA",
    "Coreia do Sul": "KOR",
    "Tchéquia": "CZE",

    "Canadá": "CAN",
    "Bósnia e Herzegovina": "BIH",
    "Catar": "QAT",
    "Suíça": "SUI",

    "Brasil": "BRA",
    "Marrocos": "MAR",
    "Haiti": "HAI",
    "Escócia": "SCO",

    "Estados Unidos": "USA",
    "Paraguai": "PAR",
    "Austrália": "AUS",
    "Turquia": "TUR",

    "Alemanha": "GER",
    "Curaçao": "CUW",
    "Costa do Marfim": "CIV",
    "Equador": "ECU",

    "Holanda": "NED",
    "Japão": "JPN",
    "Suécia": "SWE",
    "Tunísia": "TUN",

    "Bélgica": "BEL",
    "Egito": "EGY",
    "Irã": "IRI",
    "Nova Zelândia": "NZL",

    "Espanha": "ESP",
    "Cabo Verde": "CPV",
    "Arábia Saudita": "KSA",
    "Uruguai": "URU",

    "França": "FRA",
    "Senegal": "SEN",
    "Iraque": "IRQ",
    "Noruega": "NOR",

    "Argentina": "ARG",
    "Argélia": "DZA",
    "Áustria": "AUT",
    "Jordânia": "JOR",

    "Portugal": "POR",
    "RD do Congo": "COD",
    "Uzbequistão": "UZB",
    "Colômbia": "COL",

    "Inglaterra": "ENG",
    "Croácia": "CRO",
    "Gana": "GHA",
    "Panamá": "PAN"
}
```

```python
# =============================================================================
# 04. CONEXÃO COM O GOOGLE DRIVE
# Monta o Google Drive e define os caminhos dos arquivos do projeto.
# =============================================================================

from google.colab import drive
drive.mount('/content/drive')

PASTA = "/content/drive/MyDrive/Album"

ARQ_STICKERS = f"{PASTA}/stickers.csv"
ARQ_NOMES = f"{PASTA}/nomes_album.csv"
```

```python
# =============================================================================
# 05. CRIAÇÃO DA BASE DE FIGURINHAS
# Gera todas as figurinhas do álbum:
# =============================================================================

dados = []


# FWC (00 a 19)

for i in range(20):
    dados.append({
        "id_album": f"FWC{i:02d}",
        "grupo": "FWC",
        "selecao": "FIFA",
        "tipo": "Especial"
    })


# Seleções

for grupo, selecoes in grupos.items():

    for selecao in selecoes:

        sigla = siglas[selecao]

        # Escudo
        dados.append({
            "id_album": f"{sigla}01",
            "grupo": grupo,
            "selecao": selecao,
            "tipo": "Escudo"
        })

        # Foto da seleção
        dados.append({
            "id_album": f"{sigla}02",
            "grupo": grupo,
            "selecao": selecao,
            "tipo": "Time"
        })

        # Jogadores
        for jogador in range(1, 19):

            dados.append({
                "id_album": f"{sigla}{jogador + 2:02d}",
                "grupo": grupo,
                "selecao": selecao,
                "tipo": "Jogador"
            })


# Coca-Cola

for i in range(1, 15):
    dados.append({
        "id_album": f"COLA{i:02d}",
        "grupo": "COCA",
        "selecao": "Coca-Cola",
        "tipo": "Especial"
    })
```

```python
# =============================================================================
# 06. CRIAÇÃO DO DATAFRAME PRINCIPAL
# Converte a lista de registros em um DataFrame Pandas.
# =============================================================================

df = pd.DataFrame(dados)

print(f"Total de figurinhas: {len(df)}")
```

```python
# =============================================================================
# 07. EXPORTAÇÃO DA BASE PRINCIPAL
# Salva a base completa das figurinhas.
# =============================================================================

df.to_csv(ARQ_STICKERS, index=False)
```

```python
# =============================================================================
# 08. CRIAÇÃO DA BASE DE NOMES
# Gera um arquivo auxiliar para preenchimento manual dos nomes
# das figurinhas.
# =============================================================================

df_nomes = df[["id_album"]].copy()

df_nomes["nome"] = ""

df_nomes.to_csv(ARQ_NOMES, index=False)
```
